Dzielimy zmienne na ilościowe i jakościowe.
Usuwamy zmienne, które nic nie wnoszą do analizy tj. 'StandardHours', 'EmployeeCount', 'EmployeeNumber', 'Over18'  
Zmieniam typ danych w kolumnach Attrition oraz OverTime

```python
df = df.drop(columns=['StandardHours', 'EmployeeCount', 'EmployeeNumber', 'Over18'])

df['Attrition'] = df['Attrition'].replace({'Yes': 1, 'No': 0})
df['OverTime'] = df['OverTime'].replace({'Yes': 1, 'No': 0})

ilosciowe = [c for c in df.columns if df[c].dtype=='int64']
jakosciowe = [c for c in df.columns if c not in ilosciowe]

print(f'Zmienne jakościowe: {jakosciowe}')
print(f'Zmienne ilościowe: {ilosciowe}')
```

Zmieniamy typy danych w kolumnach nieliczbowych na typ category

```python
nieliczbowe_kolumny = df.select_dtypes(exclude=['number'])

# Wyświetlenie unikalnych wartości dla każdej kolumny nieliczbowej
for kolumna in nieliczbowe_kolumny.columns:
    print(f"Unikalne wartości w kolumnie '{kolumna}': {nieliczbowe_kolumny[kolumna].unique()}")
    df[kolumna] = df[kolumna].astype('category')

print('\nDla podanych kolumn, typy zostały zmienione na category')
```

#### Analiza wieku pracowników

Tworzę dwa oddzielne zbiory dla aktualnych i pracowników, którzy odeszli

```python
df_byli_pracownicy = df[df['Attrition']==1]
df_byli_pracownicy.shape
```

```python
df_aktualni_pracownicy = df[df['Attrition']==0]
df_aktualni_pracownicy.shape
```

```python
bins = np.arange(15, 61, 5) # ustawiam biny tak, żeby słupki się na siebie nakładały i pasowały do przedziałów na skali osi x

sns.histplot(df_byli_pracownicy['Age'], bins=bins, color='#c0392b', label='Byli pracownicy', alpha=0.6, kde=True)
sns.histplot(df_aktualni_pracownicy['Age'], bins=bins, color='#4f6fdf', label='Aktualni pracownicy', alpha=0.6, kde=True)

plt.title("Rozkłady wieku byłych i obecnych pracowników", fontsize=14)
plt.xlabel("Wiek", fontsize=12)
plt.ylabel("Ile osób", fontsize=12)
plt.legend()
plt.show()
```

Wśród obecnych pracowników najwięcej jest osób w wieku 30-40 lat. Łącznie około 500 osób.  
Wśród pracowników, którzy odeszli najwięcej jest osób w wieku 25-35 lat. Możemy zatem wnioskować, że jeżeli ktoś decyduje się opuścić firmę, robi to wczesniej.
Potwierdzają to mediana i średnia wieku

```python
aktualni_mediana = df_aktualni_pracownicy['Age'].median()
byli_mediana = df_byli_pracownicy['Age'].median()
aktualni_avg= round(df_aktualni_pracownicy['Age'].mean(),2)
byli_avg= round(df_byli_pracownicy['Age'].mean(),2)
print(f'Mediana i średnia wieku dla aktualnych pracowników: {aktualni_mediana} {aktualni_avg}')
print(f'Mediana i średnia wieku dla byłych pracowników: {byli_mediana} {byli_avg}')
```

#### Macierz korelacji

```python
# Tworzę macierz korelacji tylko dla zmiennych liczbowych
corr = df.corr(numeric_only=True)

# Tworzę maskę dla górnego trójkąta macierzy
mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr,
    mask=mask,             # Maskuję górny trójkąt
    annot=True,            # Wyświetlanie wartości korelacji
    annot_kws={"size": 10},  # Ustawienie wielkości czcionki dla wartości
    cmap='coolwarm',       # Kolor reprezentujący siłę korelacji
    fmt=".2f",             # Formatowanie liczb korelacji
    cbar_kws={'shrink': 0.8},  # Zmniejszenie rozmiaru paska kolorów
    linewidths=0.5
)

plt.title("Macierz Korelacji - Wszyscy pracownicy", fontsize=16)
plt.show()
```

Z danych wynika, że silnie skorelowane są JobLevel z MonthlyIncome oraz TotalWorkingYears, PerformanceRating z PercentSalaryHike.  
Warto zauważyć silną korelację pmmiędzy YearsInCurrentRole oraz YearsInCompany, co oznacza, że pracownicy długo pracują bez zmian stanowisk.  
Do tego można zauważyć dość wyraźną korelację pomiędzy tymi dwoma zmiennymi ze zmienną YearsSinceLastPromotion, co oznacza, że pracownicy mogą pracować bez awansu

```python
corr_1 = df['JobLevel'].corr(df['MonthlyIncome'])
corr_2 = df['JobLevel'].corr(df['TotalWorkingYears'])
corr_3 = df['PerformanceRating'].corr(df['PercentSalaryHike'])
corr_4 = df['YearsInCurrentRole'].corr(df['YearsAtCompany'])

print(f'Korelacja JobLevel z MonthlyIncome: {corr_1}')
print(f'Korelacja JobLevel z TotalWorkingYears: {corr_2}')
print(f'Korelacja PerformanceRating z PercentSalaryHike: {corr_3}')
print(f'Korelacja YearsInCurrentRole z YearsAtCompany: {corr_4}')
```

```python
# Tworzenie macierzy korelacji tylko dla zmiennych liczbowych
corr2 = df_byli_pracownicy.corr(numeric_only=True)

# Tworzenie maski dla górnego trójkąta macierzy
mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
# Zliczam po departamentach i liczbę odejść na poszczególnych stanowiskach
attrition_by_dept_and_job = (
    df.groupby(['Department', 'JobRole'])['Attrition']
    .sum()
    .reset_index()
)

# Filtruje wartości dodatnie, sortuje je malejąco
attrition_by_dept_and_job = attrition_by_dept_and_job[attrition_by_dept_and_job['Attrition'] > 0]
attrition_by_dept_and_job = attrition_by_dept_and_job.sort_values(by='Attrition', ascending=False)

print("Liczba odejść w poszczególnych departamentach i na stanowiskach:")
print(attrition_by_dept_and_job)

# Podsumowanie odejść w poszczególnych departamentach
attrition_by_department = df.groupby('Department')['Attrition'].sum().reset_index()
attrition_by_department = attrition_by_department[attrition_by_department['Attrition'] > 0]
attrition_by_department = attrition_by_department.sort_values(by='Attrition', ascending=False)

print("\nPodsumowanie liczby odejść w poszczególnych departamentach:")
print(attrition_by_department)
```

Jak pokazuje powyższa analiza, najwięcej pracowników odeszło z działu R&D. Dużo odejść było również w dziale sprzedażowym (Sales)
