```python
# Ścieżka do oryginalnego pliku
input_file = r"D:\Pozostałe\Infoshare\0000_Homework\01_Oil_DB\olist_order_reviews_dataset.csv"
output_file = r"D:\Pozostałe\Infoshare\0000_Homework\01_Oil_DB\olist_order_reviews_clean.csv"

# Wczytanie pliku CSV z Pandas
# Używamy engine='python' i quotechar='"', żeby Pandas radził sobie z niepełnymi cudzysłowami
df = pd.read_csv(input_file, delimiter=",", quotechar='"', engine="python", encoding="utf-8")

# Funkcja czyszcząca komentarze – jeśli zawierają przecinek, a nie są w cudzysłowie, dodajemy je
def fix_comment(text):
    if pd.isna(text):
        return text
    text = str(text)
    # Jeśli zawiera przecinek i nie zaczyna się/kończy cudzysłowem
    if "," in text and not (text.startswith('"') and text.endswith('"')):
        return f'"{text}"'
    return text

# Poprawa kolumn komentarzy
df["review_comment_message"] = df["review_comment_message"].apply(fix_comment)
df["review_comment_title"] = df["review_comment_title"].apply(fix_comment)

# Zapis poprawionego pliku
df.to_csv(output_file, index=False, encoding="utf-8", quoting=1)  # quoting=1 => QUOTE_ALL

print("Plik został zapisany jako:", output_file)
```

```python
import pandas as pd
import csv

# Wczytanie pliku CSV
df = pd.read_csv(r"D:\Pozostałe\Infoshare\0000_Homework\01_Oil_DB\olist_order_reviews_final.csv", encoding="utf-8")

# Funkcja czyszcząca: usuwa ", nowe linie i przecinki
def clean_text(text):
    if pd.isna(text):
        return text
    return (
        str(text)
        .replace('"', '')   # usuń wszystkie cudzysłowy
        .replace('\n', '')  # usuń znaki nowej linii
        .replace('\r', '')  # usuń znaki powrotu karetki
        .replace(',', '')   # usuń przecinki
    )

# Przetworzenie wskazanych kolumn
for col in ["review_comment_message", "review_comment_title"]:
    if col in df.columns:
        df[col] = df[col].apply(clean_text)

# Zapis do nowego pliku CSV
df.to_csv(r"D:\Pozostałe\Infoshare\0000_Homework\01_Oil_DB\olist_order_reviews_final_2.csv", index=False, encoding="utf-8", quoting=csv.QUOTE_MINIMAL)
```
